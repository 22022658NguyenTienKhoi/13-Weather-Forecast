```python
import pandas as pd
```

```python
weather = pd.read_csv("./hanoiweather.csv", index_col="datetime")
```

```python
weather
```

```python
weather.info()
```

```python
data_weather = weather.dropna(axis='columns',thresh=12000)
data_weather = data_weather.dropna()
data_weather.drop(data_weather.columns[9:11], axis=1, inplace=True)
data_weather.drop(data_weather.columns[0], axis=1, inplace=True)
data_weather.drop(data_weather.columns[12:], axis=1, inplace=True)
data_weather.info()
```

```python
data_weather
```

```python
data_weather[["tempmax", "tempmin"]].loc['2020-01-01':].plot(figsize=(16,8))
```

```python
data_weather["prediction"] = data_weather.shift(-1)["tempmax"]
data_weather.iloc[-1,-1] = data_weather.iloc[-1, 0]
data_weather
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.5)
```

```python
elements_predict = ["tempmax", "tempmin", "temp", "humidity"]
```

```python
train = data_weather.loc[:"2015-12-31"]
test = data_weather.loc["2016-01-01":]
```

```python
reg.fit(train[elements_predict], train["prediction"])
```

```python
predictions = reg.predict(test[elements_predict])
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(test["prediction"], predictions)
```

```python
combined = pd.concat([test["prediction"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
```
